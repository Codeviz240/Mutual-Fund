```python
def calculator(investment, r, years):
    #Returns Calculator

    months = years * 12

    #Compound Interest Formula
    i = r/100/12

    #Seperate Calculation for FV
    a = (1+i)**months-1
    b = (1+i)/i
    
    #Final Future Value Calculation
    FV = investment * a * b
    #print(FV)
    
    return FV
```
